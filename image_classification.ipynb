```python
independent=data.iloc[:,1:]
target=data.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,target,test_size=0.3)
```

```python
x_train
```

```python
rf=DecisionTreeClassifier()
rf.fit(x_train,y_train)
```

```python
pred=rf.predict(x_test)
```

```python
pred
```

```python
s=y_test.values
count=0
for i in range (len(pred)):
    if pred[i]==s[i]:
        count+=1
print(count)
```
